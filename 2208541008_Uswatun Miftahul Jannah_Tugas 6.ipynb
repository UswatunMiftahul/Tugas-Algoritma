5.21.4 Mengubah data pada Pandas

```python
data_csv['luas sepal'] = data_csv['sepal length (cm)'] * data_csv['sepal width (cm)']
print(data_csv)
#membuat data baru yaitu kolom luas sepal dengan ketentuan data sepal length dikali dengan data sepal width
```

```python
#menghilangkan kolom luas sepal dari dataframe
data_csv = data_csv.drop(columns=['luas sepal'])
print(data_csv)
```

```python
#membuat data baru sesuai dengan ketentuan
#menampilkan kolom jumlah lebar dengan ketentuan data lebar sepal ditambahkan data lebar petal
data_csv['jumlah lebar'] = data_csv['sepal width (cm)'] + data_csv['petal width (cm)']
print(data_csv)
```

```python
#menghilangkan kolom jumlah lebar dari dataframe
data_csv = data_csv.drop(columns=['jumlah lebar'])
print(data_csv)
```

```python
#fungsi ini adalah memilih data dari dataframe
#dengan (:) memilih semua baris, dan (1:3) yaitu dari kolom 1 hingga kolom 3
#lalu sum(axis=1) adalah menggunakan sum untuk setiap baris
data_csv['sepal'] = data_csv.iloc[:,1:3].sum(axis=1)
print(data_csv)
#sehingga akan menampilkan kolom baru yaitu sepal yang berisi hasil penjumlahan dari kedua kolom pada subset dataframe yg telah dipilih
```

```python
data_csv['petal width (cm)'] = data_csv['petal width (cm)'].replace([0.2],'none')
print(data_csv)
#mengubah isi kolom petal width yang berisi angka 0.2 menjadi none pada dataframe
```

5.22 MATPLOTLIB
adalah Library pada Python yang fungsi utamanya
adalah dalam hal menggambarkan grafik, histogram dan berbagai jenis plot hasil
dari pengolahan data menggunakan Library lainnya yaitu Pandas dan Numpy

```python
#hal pertama yang dilakukan adalah kita memanggil library matplotlib
import matplotlib.pyplot as plt
```

Grafik Garis

```python
y = [45, 50, 55, 60, 68]
x = [2015, 2016, 2017, 2018, 2019]
plt.plot(x,y)

plt.xlabel('Tahun')
plt.ylabel('Jumlah siswa')

plt.title('Lulusan SMA 1')
plt.grid(True)

plt.show()

#pertama kita tentukan dulu angka yang berada dalam variabel x dan y
#lalu kita menggunakan plt.plot(x,y) untuk meletakkan posisi titik di x dan y
#lalu kita memberikan keterangan pada sumbu x(horizontal) menggunakan plt.xlabel
#dan memberikan keterangan pada sumbu y(vertikal) menggunakan plt.ylabel
#selanjutnya kita bisa memberikan judul grafik tersebut menggunakan plt.title
#selanjutnya kita gunakan plt.grid untuk menampilkan garis grid
#dan kita menggunakan plt.show untuk menampilkan chart
```

Grafik Batang

```python
x = [1, 2, 3, 4, 5]
y = [80, 90, 95, 85, 79]

fig, ax = plt.subplots()

ax.bar(x, y, color= 'tan')

ax.set_title('Nilai Kelas A')
ax.set_ylabel('Nilai')
ax.set_xlabel('Mata Pelajaran')

ax.set_xticks(x)
ax.set_xticklabels(("Ips", "Ipa", "Bahasa", "Pkn", "Agama"))

plt.show()

#pertama kita tentukan dulu angka yang berada dalam variabel x dan y
#lalu kita gunakan fig,ax untuk menambahkan kolom figure dan axes pada sublplots()
#kemudian kita gunakan ax.bar untuk membuat output persegi
#lalu kita gunakan ax.set_title untuk membuat judul grafik
#lalu kita gunakan ax.set_ylabel untuk menamai garis vertikal dan ax.set_xlabel untuk menamai garis horizontal
#lalu kita gunakan ax.set_xticks untuk memberikan nilai pada garis horizontal
#kemudian ada ax.set_xticklabels untuk menamai nilai yang diciptakan xticks
#lalu terakhir kita gunakan plt.show untuk menampilkan chart
```

Diagram Lingkaran

```python
labels = ['Dokter', 'Guru', 'Pilot', 'Polisi']
quantity = [7, 9, 8, 11]
colors = ['aquamarine', 'plum', 'grey', 'lightcoral']

plt.title('Daftar Cita-cita Anak SD 4')
plt.pie(quantity, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)

plt.axis('equal')
plt.show()

#pertama kita tentukan dulu isi label untuk setiap data
#kemudian kita gunakan quantity untuk menentukan kuantitas dari setiap label
#setelah itu kita gunakan colors untuk menentukan warna nya
#lalu kita gunakan plt.title untuk membuat judul grafik
#lalu kita gunakan plt.pe merupakan fungsi untuk mengidentifikasi label, kuantitas, dan warna, shadow digunakan untuk memberikan efek bayangan, startangle=90 digunakan untuk membuat tabel pada nilai pertama senilai 90 derajat
#setelah itu plt.axis untuk memberikan aspek rasio pada setiap nilai
#terakhir plt.show untuk menampilkan grafik
```

```python
fig, axs = plt.subplots(1, 1, figsize=(12, 8))

axs.scatter(data_csv['petal width (cm)'], data_csv['sepal width (cm)'], color="skyblue", label="Width")
axs.scatter(data_csv['petal length (cm)'], data_csv['sepal length (cm)'], color="violet", label="Length")
plt.title('Perbandingan panjang dan lebar pada sepal dan petal')

plt.xlabel('Petal')
plt.ylabel('Sepal')

axs.legend()

plt.show()

#pertama kita membuat subplots dengan 2 baris dan 1 kolom
#lalu membuat sebuah plot dengan axs.scatter untuk menentukan properti pembentuk bagan
#lalu kita membuat judul menggunakan plt.title
#lalu kita gunakan plt.ylabel untuk menamai garis vertikal dan plt.xlabel untuk menamai garis horizontal
#lalu kita menggunakan axs.legend untuk titik yang merupakan gabungan x dan y
#terakhir kit gunakan plt.show untuk menampilkan plot
```
